```python
import requests,pathlib,zipfile,io
def Desc():
    print("Downloading ESC-50")
    buf = requests.get("https://github.com/karolpiczak/ESC-50/archive/master.zip").content
    zipfile.ZipFile(io.BytesIO(buf)).extractall(".")
    pathlib.Path("ESC-50-master").rename("ESC-50")
    return "Success"
print(Desc())
```
